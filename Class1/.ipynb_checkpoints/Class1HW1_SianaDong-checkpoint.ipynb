```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
state = pd.read_csv("state.csv")
state.head()
```

```python
state.shape
```

```python
state.dtypes
```

```python
state.describe()
```

```python
freq_state = pd.crosstab(index=state["State"], columns="count")

freq_state
```

```python
plt.bar(state["State"], freq_state["count"])
plt.xticks(rotation='vertical')
plt.show()
```

```python
freq_abbrev = pd.crosstab(index=state["Abbreviation"], columns="count")

freq_abbrev
```

```python
plt.bar(state["Abbreviation"], freq_abbrev["count"])
plt.xticks(rotation='vertical')
plt.show()
```

```python
plt.bar(state["Abbreviation"], state["Murder.Rate"])
plt.xticks(rotation='vertical')
plt.show()
```

```python
state.plot.scatter(x='Population', y='Murder.Rate')
```
